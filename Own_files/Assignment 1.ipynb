```python
# Part I
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression

traffic_density = np.array([50,20,40,10,5])
traffic_volume = np.array([2000,900,1500,500,300])
is_congested = np.array([1,0,1,0,0])

df = pd.DataFrame({'traffic_density': traffic_density, 'traffic_volume': traffic_volume, 'is_congested': is_congested})
clf = LogisticRegression()
clf = clf.fit(df[['traffic_density','traffic_volume']],df['is_congested'])

intercept = clf.intercept_[0]
coeff1 = clf.coef_[0][0]
coeff2 = clf.coef_[0][1]

df['discriminant_function_output'] = intercept + coeff1 * df['traffic_density'] + coeff2 * df['traffic_volume']
print(df)

print('The features of the model are the traffic_density and traffic_volume')
print('The dependent variable is "Evaluated as Congested", indicating whether the road is congested or not')
print(f'the discriminant function is = {clf.intercept_[0]} + {clf.coef_[0][0]} * traffic_density + {clf.coef_[0][1]} * traffic_volume' )
```

```python
# Part II
def discriminant_function(intercept, coeff1, coeff2, df_tmp):
    df_tmp['discriminant_function_output'] = intercept + coeff1 * df_tmp['traffic_density'] + coeff2 * df_tmp['traffic_volume']
    return df_tmp

def find_probability(df_tmp):
    df_tmp['probability'] = 1 / (1 + np.exp(-df_tmp['discriminant_function_output']))
    return df_tmp

pred = clf.predict(np.array([30,1000]).reshape(1,-1))
df_part2 = pd.DataFrame({'traffic_density': np.array([30]), 'traffic_volume': np.array([1000])})

df_part2 = discriminant_function(clf.intercept_[0], clf.coef_[0][0],clf.coef_[0][1], df_part2)
df_part2 = find_probability(df_part2)
print(f"The discriminant_function is {df_part2['discriminant_function_output'].values[0]}")
print(f"The probability is {df_part2['probability'].values[0]}")
print(f'Therefore the prediction for whether it is congested is {pred[0]}, that means it is not congested')
```
